```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
dataset = pd.read_csv('./dataset/KPMG Dataset.csv')

dataset.head()
```

```python
df = dataset[dataset['order_status'] == 'Approved']
```

```python
# Check for null values in each column
nulls_per_column = df.isnull().sum()

# Display null counts per column
nulls_per_column
```

```python
# Count the occurrences of duplicate rows
num_duplicates = df.duplicated().sum()

if num_duplicates > 0:
    # Drop duplicate rows
    df.drop_duplicates(inplace=True)
    print(f"Dropped {num_duplicates} duplicate rows.")
else:
    print("No duplicates found.")
```

```python
# Drop record having null value in customer_id
df = df.dropna(subset=['customer_id'])
```

```python
df.dtypes
```

```python
# Convert to datetime
df['transaction_date'] = pd.to_datetime(df['transaction_date'], format='%d/%m/%Y')
```

```python
df.dtypes
```

```python
# Convert to first day of month of the given date
def get_month(x): return dt.datetime(x.year, x.month, 1)

# Get order month
df['order_month'] = df['transaction_date'].apply(get_month)
```

```python
# Get corhort month
df_first_order_month_by_customer = df.groupby('customer_id')['order_month'].min().reset_index()
df_first_order_month_by_customer.columns = ['customer_id', 'cohort_month']
```

```python
# Add cohort month to current df
df = pd.merge(df, df_first_order_month_by_customer, on='customer_id', how='left')
```

```python
# Define a function to get cohort index
def get_cohort_index(row, order_month_col, cohort_col):
    year_diff = row[order_month_col].year - row[cohort_col].year
    month_diff = row[order_month_col].month - row[cohort_col].month
    return year_diff * 12 + month_diff + 1
```

```python
# Add cohort index column
df['cohort_index'] = df.apply(get_cohort_index, axis=1, args=('order_month', 'cohort_month'))
```

```python
# Create a pivot table
pivot_df = df.pivot_table(index='cohort_month', columns='cohort_index', values='customer_id', aggfunc=pd.Series.nunique)
pivot_df.index = pivot_df.index.strftime('%Y-%m')
pivot_df
```

```python
# Calculate retention rate
retention_rate = pivot_df.divide(pivot_df.iloc[:, 0], axis=0)
retention_rate
```

```python
# Create a heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(pivot_df, annot=True, fmt='.0f', cmap='YlGnBu')  # 'annot' adds values, 'fmt' formats the annotations

plt.title('Cohort Analysis Heatmap', y=-0.01)
plt.xlabel('Cohort Index')
plt.ylabel('Cohort Month')
plt.yticks(rotation=0)

plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top') 

plt.show()
```

```python
# Create a heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(retention_rate, annot=True, fmt='.1%', cmap='YlGnBu')  # 'annot' adds values, 'fmt' formats the annotations

plt.title('Cohort Analysis Heatmap', y=-0.01)
plt.xlabel('Cohort Index')
plt.ylabel('Cohort Month')
plt.yticks(rotation=0)

plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top') 

# Alternative way to add %
# ax = sns.heatmap(retention_rate, annot=True, fmt='.1%', cmap='YlGnBu')  # 'annot' adds values, 'fmt' formats the annotations
# for t in ax.texts:
#     t.set_text(t.get_text() + " %")

plt.show()
```
